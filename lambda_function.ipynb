Lambda Square

```python
number_list=[2,55,47,66,3,6,8]

square_list= map(lambda x: x**2, number_list)

print(list(square_list))
```

```python
def square(x):
    return x*x
y=[2,4,5,6]
square_number=map(square,y)

print(list(square_number))
```
