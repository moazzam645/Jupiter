pass keyword

```python
for i in range(5):
    if i ==3:
        pass
    print(i)
```

Continue Keyword


```python
for i in range(5):
    if i ==3:
        continue
    print(i)
```

Break Keyword

```python
for i in range(5):
    if i ==3:
        break
    print(i)
```

while 

```python

i=1
while i in range(5):
    print(i)
    i=i+2
    
```

Try, Except and Finally

```python
try:
    print("lets check division")
    print(4/0)
except ZeroDivisionError:
    print("zero division error fix it")
finally:
    print("it will run no matter what happens")
```
